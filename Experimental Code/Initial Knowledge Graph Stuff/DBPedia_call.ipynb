```python
def check_value_exists(initial_dic):
    original_names=[]
    vals=[]
    labels=[]
    column_names=initial_dic.keys()
    for name in tqdm(column_names):
        original_name=name
        name=name.lower()
        try:
            name=name.replace("&", "and")
        except:
            continue
        call_list=name.split(" ")
        count=0
        call='NA'
        for item in call_list:
            doc = nlp('{}'.format(item))
            if len([(ent.text, ent.kb_id_, ent._.dbpedia_raw_result['@similarityScore']) for ent in doc.ents])==0:
                continue
            else:
                print('here')
                call=doc.ents[0].kb_id_
                count+=1
        
        if count==0:
            vals.append('N')
            labels.append(call)
            original_names.append(original_name)

        else:
            vals.append('Y')
            labels.append(call)
            original_names.append(original_name)        

    dic={'IAB_Label':original_names, 'Label': labels,"Value": vals}

    df=pd.DataFrame.from_dict(dic)

    print(df['Value'].value_counts())
    
    return df
        
        
```

```python
def DBPediaCall(initial_dic,metalist):
    column_names=initial_dic.keys()
    for name in tqdm(column_names):
        original_name=name
        name=name.lower()
        try:
            name=name.replace("&", "and")
        except:
            continue
        doc = nlp('{}'.format(name))
        if len([(ent.text, ent.kb_id_, ent._.dbpedia_raw_result['@similarityScore']) for ent in doc.ents])!=0:
            try:
                print('in spacy')
                call=doc.ents[0].kb_id_.split('/')[-1]
                obj=requests.get("https://dbpedia.org/data/{}.json".format(call)).json()
                keys=obj.keys()
                for key in keys:
                    if 'resource' in key:
                        temp=[key.split('/')[-1], 'is_related_to', original_name]
                        metalist.append(temp)
                        
            except Exception as e:
                print(e)
                pass
        else:

            name_list=name.split(' and ')
            for k,v in enumerate(name_list):
                p = inflect.engine()
                if v == 'business':
                    name_list[k]=v 
                    continue

                if p.singular_noun(v)==False:
                    name_list[k]=v 
                else:
                    name_list[k]=p.singular_noun(v)
                
                obj=requests.get("https://lookup.dbpedia.org/api/search/KeywordSearch?&maxResults=10000&QueryString={}&format=json_raw&minRelevance=1000".format(name_list[k])).json()
                print(name_list[k])
                if len(obj['docs'])!=0:

                    if obj['docs'][0]['resource'][0].split('/')[-1].lower() == name_list[k]:
                        try:
                            sub_obj=requests.get("https://dbpedia.org/data/{}.json".format(obj['docs'][0]['resource'][0].split('/')[-1])).json()
                            keys=sub_obj.keys()
                            print('In_if')
                            print(keys)
                            for key in keys:
                                if 'resource' in key:
                                    temp=[key.split('/')[-1], 'is_related_to', original_name]
                                    metalist.append(temp)
                        except:
                            for i in range(len(obj['docs'])):
                                key=obj['docs'][i]['resource'][0]
                                if 'resource' in key:
                                    temp=[key.split('/')[-1], 'is_related_to', original_name]
                                    metalist.append(temp)

                    else:
                        for i in range(len(obj['docs'])):
                            key=obj['docs'][i]['resource'][0]
                            if 'resource' in key:
                                temp=[key.split('/')[-1], 'is_related_to', original_name]
                                metalist.append(temp)
                else:
                    continue

    return metalist
                
            

            
            
        
        
    
```
