```python
import requests
import pandas as pd
import pickle as pkl
import csv
from tqdm.auto import tqdm
from time import sleep
```

```python
global_list=[]
```

```python
MAF_KG=[]
```

```python
audio_list=[]
```

```python
audio_df=pd.read_csv('ConceptNet_MAF_Data_Files/audio_classes.csv')
```

```python
audio_list=audio_df['display_name'].to_list()
```

```python
print(audio_list)
```

```python
for k,audio in enumerate(audio_list):
    audio=audio.lower()
    audio=audio.split(', ')
    bruh=[]
    for items in audio:
        try:
            items=items.replace(" ", "_")
            bruh.append(items)
        except:
            bruh.append(items)
            pass

    audio_list[k]=bruh
```

```python
audio_list
```

```python
for items in audio_list:
    for item in items:
        global_list.append(item)
        temp=[item, 'is_related_to', 'Audio']
        print(temp)
        MAF_KG.append(temp)
```

```python
len(MAF_KG)
```

```python
def conceptNet_call(list_of_lists, MAF_KG):
    for items in tqdm(list_of_lists):
        for item in items:
            obj = requests.get("http://api.conceptnet.io/query?node=/c/en/{}&limit=100000".format(item)).json()
            if len(obj['edges'])==0:
                continue
            for edge in obj['edges']:
                if edge['rel']['label']=='ExternalURL':
                    continue
                elif edge['end']['term']=='/c/en/{}'.format(item):
                    if edge['start']['language'] == 'en':
                        temp=[edge['start']['term'].split('/')[-1], 'is_related_to', item]
                        print(temp)
                        MAF_KG.append(temp)
                elif edge['start']['term']=='/c/en/{}'.format(item):
                    if edge['end']['language'] == 'en':
                        temp=[edge['end']['term'].split('/')[-1], 'is_related_to', item]
                        print(temp)
                        MAF_KG.append(temp)

    return MAF_KG
```

```python
print(audio_list)
```

```python
MAF_KG=conceptNet_call(audio_list,MAF_KG)
```

```python
places_file=open('ConceptNet_MAF_Data_Files/categories_places365.txt')
places_list = places_file.readlines()
print(places_list)
```
