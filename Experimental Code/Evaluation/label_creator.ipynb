```python
H_IAB_words=df_unclean_only_H['adword'].tolist()
```

```python
H_IAB_words
```

```python
child_to_parent=defaultdict(list)
H_IAB_words_set=list(set(H_IAB_words))
for adwords in H_IAB_words_set:
    H_IAB=adwords.split('|')[-1]
    vals=adwords.split('|')
    for ad in vals:
        if len(ad) != 0 and ad != H_IAB and ad not in child_to_parent[H_IAB]:
            child_to_parent[H_IAB].append(ad)
```

```python
child_to_parent
```

```python
with open('child_parent_dic.pkl', 'wb') as f:
    pkl.dump(child_to_parent, f)
```

```python
new_MAF_words=[]
IAB_labels=[]
for items,H_IAB in zip(MAF_words,H_IAB_words):
    H_IAB=H_IAB.split('|')[-1]
    if type(items) is list:
        # print('here')
        # print(items)
        # new_MAF_words=list(map(str,items[1:-1].split(',')))
        # MAF_words[k]=new_MAF_words
        for mini in items:
            new_MAF_words.append(mini.strip())
            IAB_labels.append(H_IAB)

    else:
        new_MAF_words.append(items.strip())
        IAB_labels.append(H_IAB)

        
```

```python
len(new_MAF_words)
```

```python
len(IAB_labels)
```

```python
dic=defaultdict(list)
```

```python
for maf, iab in zip(new_MAF_words,IAB_labels):
    dic[maf].append(iab)
```

```python
len(dic.keys())
```

```python
for k,v in dic.items():
    print( k, v)
```

```python
new_dic={'item':dic.keys(),'IAB_Candidates':dic.values()}
df=pd.DataFrame.from_dict(new_dic)
scores=[3]*1876
df['scores']=scores
df.to_csv("MAF_Evaluation_New_mapping.csv")
```
